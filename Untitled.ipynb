## Defining free email domains

```python
free_email_domains = [
    "gmail.com",
    "yahoo.com",
    "hotmail.com",
    "outlook.com",
    "icloud.com",
    "aol.com",
    "protonmail.com"
]
```

## creating Email classification function 

```python
def classify_email(email):
    if "@" not in email or "." not in email:
        return "invalid"

    domain = email.split("@")[-1]

    if domain in free_email_domains:
        return "free_email"

    return "business_email"
```

## Applying the classification to the dataset

```python
df["email_type"] = df["email"].apply(classify_email)
```

## Lead scoring logic

## Defineing target countries

```python
target_countries = [
    "United States",
    "Canada",
    "United Kingdom",
    "Germany",
    "Australia"
]
```

## Lead scoring function 

```python
def calculate_lead_score(row):
    score = 0

    # Email quality
    if row["email_type"] == "business_email":
        score += 40
    elif row["email_type"] == "free_email":
        score += 10

    # Company present
    if pd.notna(row["company"]) and row["company"].strip() != "":
        score += 20

    # Lead source quality
    if "linkedin" in row["lead_source"].lower():
        score += 20

    # Target country
    if row["country"] in target_countries:
        score += 20

    return score
```

## Applying scoring to the dataset

```python
df["lead_score"] = df.apply(calculate_lead_score, axis=1)
```

## CRM ready outputs

## Assigning lead priority labels

```python
def assign_priority(score):
    if score >= 70:
        return "High"
    elif score >= 40:
        return "Medium"
    return "Low"

df["lead_priority"] = df["lead_score"].apply(assign_priority)
```
